# DATA CLEANING IN PANDAS

# QUESTION: Use Pandas to clean and preprocess a messy dataset, documenting the steps taken during during the cleaning process.

## 1. Importing pandas library

```python
# importing pandas library
import pandas as pd
```

## 2. Importing dataset from excel

```python
# importing dataset from excel
df = pd.read_excel (r"C:\Users\USER\Documents\Data set 3mtt.xlsx")
```

## 3. Displaying dataset

```python
# Displaying dataset
df
```

## 4. Dropping duplicates in dataset

```python
# dropping duplicates 
df = df.drop_duplicates()
df
```

## 5. Dropping not useful columns

```python
# dropping columns
df = df.drop(columns = "Address")
df
```

## 6. Stripping

```python
# stripping
df["Surname"] = df["Surname"].str.strip("123._")
df["State_of_Residence"] = df["State_of_Residence"].str.strip("/")
df
```

## 7.Cleaning and Standardizing phone numbers

```python
# cleaning and standardizing phone numbers
df["Phone_number"] = df["Phone_number"].str.replace('-', '')
df
```

## 8. Filling Null Values

```python
# filling Null Values
df = df.fillna('')
df
```

## 9. Inconsistent values

```python
# Keeping the columns consistent
df["Gender"] = df["Gender"].str.replace('Female', 'F')
df["Gender"] = df["Gender"].str.replace('Male', 'M')
df
```

## 10. Filtering down Rows of the dataset

```python
# dropping null values

for x in df.index:
    if df.loc[x, "Phone_number"] == '':
        df.drop(x, inplace=True)
        
for x in df.index:
    if df.loc[x, "Education_level"] == '':
        df.drop(x, inplace=True)        
df
```

## 11. Resetting Index

```python
df.reset_index()
```

```python
df = df.reset_index(drop=True)
df
```
